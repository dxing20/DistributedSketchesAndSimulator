```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

superspreader_true_10million = pd.read_csv("../superspreaders_true_10.csv")
superspreader_true_8million = pd.read_csv("../superspreaders_true_8.csv")
superspreader_true_6million = pd.read_csv("../superspreaders_true_6.csv")
superspreader_true_4million = pd.read_csv("../superspreaders_true_4.csv")
superspreader_true_2million = pd.read_csv("../superspreaders_true_2.csv")

# Calculate rank and log_rank for each dataset
superspreader_true_10million["rank"] = superspreader_true_10million["freq"].rank(ascending=False)
superspreader_true_10million["log_rank"] = np.log10(superspreader_true_10million["rank"])

superspreader_true_8million["rank"] = superspreader_true_8million["freq"].rank(ascending=False)
superspreader_true_8million["log_rank"] = np.log10(superspreader_true_8million["rank"])

superspreader_true_6million["rank"] = superspreader_true_6million["freq"].rank(ascending=False)
superspreader_true_6million["log_rank"] = np.log10(superspreader_true_6million["rank"])

superspreader_true_4million["rank"] = superspreader_true_4million["freq"].rank(ascending=False)
superspreader_true_4million["log_rank"] = np.log10(superspreader_true_4million["rank"])

superspreader_true_2million["rank"] = superspreader_true_2million["freq"].rank(ascending=False)
superspreader_true_2million["log_rank"] = np.log10(superspreader_true_2million["rank"])

# Plot all curves on one figure
plt.figure(figsize=(10, 6))
plt.scatter(superspreader_true_10million["log_rank"], superspreader_true_10million["freq"], color="blue", label="10 million")
plt.scatter(superspreader_true_8million["log_rank"], superspreader_true_8million["freq"], color="green", label="8 million")
plt.scatter(superspreader_true_6million["log_rank"], superspreader_true_6million["freq"], color="red", label="6 million")
plt.scatter(superspreader_true_4million["log_rank"], superspreader_true_4million["freq"], color="orange", label="4 million")
plt.scatter(superspreader_true_2million["log_rank"], superspreader_true_2million["freq"], color="purple", label="2 million")

plt.xlabel("Rank of Sources (log10)")
plt.ylabel("Spread")
plt.title("True Superspreader Frequency vs Rank")
plt.legend()
plt.show()
```
